Extracting data from the IMDB website with web scraping framework beautiful soup.
Python's package for parsing HTML documents.

Following information will be collecting from 'IMDB top 250 list':
* Movie Title- Movie name;
* Year - The year movie was released;
* Rating - The IMDB rating of the movie;
* DirectorCast- Director and cast&crew.

*For convinience of review, the project has been run already.*

```python
import requests  #The requests module allows us to send HTTP requests using Python
from bs4 import BeautifulSoup as bs 
import json 
import pandas as pd  # Data analysis and manipulation tool
```

```python
url='https://www.imdb.com/chart/top/?ref_=nv_mv_250'
page=requests.get(url)
page.status_code
```

```python
soup = bs(page.content, "html.parser")  #parser  a string consisting of the name of the parser to be used; 
print(soup.prettify())
```

```python
Movies=soup.find("tbody",class_="lister-list").find_all("tr")
#y=soup.find_all("td") 
for item in Movies:
    print(item)
type(soup)
```

```python
data={}
x=soup.find_all("td", class_="titleColumn")
for element in x:
    data[element.a.get_text()]={}
for key,value in data.items():
    print(key,value)
```

```python
for element in x:
    crew=element.a["title"]
    data[element.a.get_text()]["crew"]=crew
for key,value in data.items():
    print(key,value)
```

```python
for element in x:
    year=element.span.get_text()
    data[element.a.get_text()]["year"]=year
for key,value in data.items():
    print(key,value)
```

```python
movie_rating=soup.find_all("td",{"class":"ratingColumn imdbRating"})
ratings = []
for element in movie_rating:
    rating=element.get_text()
    ratings.append(rating.strip())
    print(rating)
```

```python
#The zip() function returns a zip object, 
#which is an iterator of tuples where the first item in each passed iterator is paired together, 
#and then the second item in each passed iterator are paired together etc.
for i, r in zip(data.keys(),ratings) :
    data[i]["rating"] = r
    
```

```python
with open("imdb_top_250_movies.json","w") as writeJSON:
    json.dump(data,writeJSON)
```
